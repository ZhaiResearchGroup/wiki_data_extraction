Helper functions for printing

```python
def print_info(row):
#     print(row['normalized_score'])
    print("Title: {}".format(row['title']))
    print("Query: {}".format(row['raw_query']))
    print("Summary: {}".format(row['sentence_summary']))
    print()

def print_rows(print_df, rows=None):
    if rows:
        for i in rows:
            row = df.iloc[i]
            print_info(row)
    else:
        for row in print_df.iterrows():
#             print(row[0])
            print_info(row[1])

# print_rows(df[15:46])
```
